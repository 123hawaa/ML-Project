```python
df_final['date'] = df_final['Date_of_Journey'].str.split('/').str[0]
df_final['month'] = df_final['Date_of_Journey'].str.split('/').str[1]
df_final['year'] = df_final['Date_of_Journey'].str.split('/').str[3]
```

```python
df_final.head(2)
```

```python
df_final['date']=df_final['date'].astype(int)
df_final['month']=df_final['month'].astype(int)
```

```python
df_final.info()
```

```python
df_final.drop('Date_of_Journey',axis =1,inplace=True )
```

```python
df_final['Arrival_Time'] = df_final['Arrival_Time'].str.split(' ').str[0]
```

```python
df_final['Arrival_hour'] = df_final['Arrival_Time'].str.split(':').str[0]
df_final['Arrival_min'] = df_final['Arrival_Time'].str.split(':').str[1]
```

```python
df_final.drop('Arrival_Time',axis =1,inplace=True )
```

```python
df_final.head(2)
```

```python
df_final.info()
```

```python
df_final['Arrival_hour'] = df_final['Arrival_hour'].astype(int)
df_final['Arrival_min'] = df_final['Arrival_min'].astype(int)
```

```python
df_final.info()
```

```python
df_final['dep_hour'] = df_final['Dep_Time'].str.split(':').str[0]
df_final['dep_min'] = df_final['Dep_Time'].str.split(':').str[1]
df_final['dep_hour'] = df_final['Arrival_hour'].astype(int)
df_final['dep_min'] = df_final['Arrival_min'].astype(int)
df_final.drop('Dep_Time',axis =1,inplace=True )
```

```python
df_final.info()
```

```python
df_final['Total_Stops'].unique()
```

```python
df_final['Total_Stops'] =df_final['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':3,'4 stops':4,'nan':1})
```

```python
df_final.info()
```

```python
df_final.drop('Route',axis = 1,inplace=True)
```

```python
df_final['Duration'] = df_final['Duration'].str[0]
```

```python
df_final['Duration'] = df_final['Duration'].astype(int)
```

```python
df_final.info()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder= LabelEncoder()
```

```python
df_final['Airline']=labelencoder.fit_transform(df_final['Airline'])
df_final['Source']=labelencoder.fit_transform(df_final['Source'])
df_final['Destination']=labelencoder.fit_transform(df_final['Destination'])
df_final['Additional_Info']=labelencoder.fit_transform(df_final['Additional_Info'])
```
